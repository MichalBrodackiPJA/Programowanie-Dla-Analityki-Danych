# Wprowadzenie: istotność statystyczna i prawdopodobieństwo

Celem testowania hipotez jest pomoc różnym naukowcom w dojściu do tego samego wniosku na podstawie tych samych danych. Ale jak połączyć idee „obiektywizmu” i „przypadku”?

Tutaj przydają się prawdopodobieństwo i rozkłady prawdopodobieństwa. Ponieważ _prawdopodobieństwo_ jest miarą tego, jak coś jest możliwe, a _rozkłady prawdopodobieństwa_ mówią nam o prawdopodobieństwie/gęstości prawdopodobieństwa każdego możliwego wyniku, możemy użyć rozkładów prawdopodobieństwa, aby potwierdzić __prawdopodobieństwo, że coś jest prawdą, biorąc pod uwagę naszą hipotezę zerową__. Oznacza to, że jeśli prawdopodobieństwo zaobserwowania jakiejś __statystyki testowej__ i jakichkolwiek rzadszych wartości, które są oparte na naszych danych przy założeniu hipotezy zerowej, jest _wystarczająco małe,_ możemy ją odrzucić. Prawdopodobieństwo to jest znane jako __wartość-p__, a _próg_ prawdopodobieństwa odrzucenia hipotezy zerowej jest znany jako __poziom istotności__ ($\mathbf{\alpha}$), który jest określany w zależności od tego, jak bardzo musimy być pewni. Jeśli wartość p, którą obliczamy na podstawie naszych danych, jest mniejsza niż wybrany przez nas poziom istotności, nasze wyniki są __istotne statystycznie__ i możemy odrzucić hipotezę zerową.

__Uwaga:__ Wartość p NIE jest prawdopodobieństwem zaobserwowania statystyki testowej, ale prawdopodobieństwem zaobserwowania statystyki testowej lub jakichkolwiek _rzadszych_ wartości. Jak myślisz, dlaczego nie obliczamy prawdopodobieństwa zaobserwowania statystyki testowej?

# Rozkłady normalne, standaryzacja i testy Z

Rozkład normalny jest krzywą w kształcie dzwonu wyśrodkowaną wokół średniej, której szerokość jest określona przez jej odchylenie standardowe. Jest to użyteczny rozkład, ponieważ możemy aproksymować wiele zmiennych losowych, aby mieć rozkład normalny, a _[centralne twierdzenie graniczne](https://www.youtube.com/watch?v=_YOr_yYPytM)_, pozwala nam dokonywać użytecznych przybliżeń. Zbadamy teraz rozkład normalny w odniesieniu do testowania hipotez.

Jeśli chodzi o testowanie hipotez, często używamy rozkładu normalnego, aby uzyskać miary prawdopodobieństwa. Mamy jednak tendencję do upraszczania rozkładu normalnego, aby zawierał tylko niezbędne komponenty do testowania. Więc jakie części rozkładu normalnego faktycznie wpływają na obliczane przez nas prawdopodobieństwo? Pierwszą rzeczą, którą musimy zrozumieć, jest to, że położenie środka jest nieistotne. __Dbamy tylko o odsunięcie od środka__. Po drugie, odchylenie standardowe wpływa na prawdopodobieństwo danego przesunięcia od średniej, co oznacza, że ​​__prawdopodobieństwo zależy od przesunięcia od średniej proporcjonalnego do odchylenia standardowego.__

To daje nam motywację do stworzenia niezwykle potężnego narzędzia, znanego jako __standaryzacja__. Celem standaryzacji jest porównanie dowolnych danych o rozkładzie normalnym ze __standardowym rozkładem normalnym__, który jest rozkładem normalnym ze średnią populacji $\mu = 0$ i odchyleniem standardowym populacji $\sigma = 1$. Jest to przydatne, ponieważ nie potrzebujemy wyraźnej średniej i odchylenia standardowego, ale raczej _ile odchyleń standardowych od średniej wynosi_, aby obliczyć jej prawdopodobieństwo.

Aby ujednolicić zbiór danych, wykonujemy następujące dwa kroki:
- Odejmij średnią od wszystkich punktów w naszym zbiorze danych, aby upewnić się, że jest wyśrodkowany na 0
- Podziel wynik przez odchylenie standardowe zbioru danych, skalując wyniki tak, aby odchylenie standardowe wynosiło 1

Dla każdej wartości danych, którą standaryzujemy, oznaczamy ją tak, jak pokazano poniżej:

$$z = \frac{x-\mu}{\sigma}$$

Gdzie $x$ to nasz punkt danych, określany jako __wynik surowy__, a $z$ to __z-score__ lub __wynik standardowy__. Z definicji jest to miara tego, o ile odchyleń standardowych punkt znajduje się od średniej.

Dlaczego zawracamy sobie głowę używaniem standardowego rozkładu normalnego? Czy nie moglibyśmy uzyskać naszych prawdopodobieństw z dystrybucji _surowych_? W praktyce _można_ po prostu użyć oryginalnego rozkładu normalnego, nawet w teście statystycznym. Jednak używając statystyki Z, możemy użyć _jednego_ holistycznego rozkładu do opisania prawdopodobieństw _dowolnych_ zestawów danych o rozkładzie normalnym. Innym powodem jest to, że skoro aby uzyskać prawdopodobieństwo z gęstości zmiennej losowej, musimy znaleźć pole pod jego krzywą w danym zakresie, musielibyśmy wykonać złożone operacje dla każdego rozkładu. Zamiast tego statystycy stworzyli __z-tablicę__, która informuje nas o _skumulowanym prawdopodobieństwie_ danego z-score. Możemy to wykorzystać do obliczenia prawdopodobieństw dla dowolnego rozkładu normalnego! Chociaż nie jest to tak istotne, jak wtedy, gdy nie mogliśmy użyć specjalnych kalkulatorów i komputerów, aby podać nam te prawdopodobieństwa, nadal znacznie upraszcza to proces.

Rozważmy przykład dla wzrostu uczniów, oznaczonego przez $X$, gdzie $X \sim N(175, 4^{2})$. Dla dowolnego ciągłego rozkładu prawdopodobieństwa możemy przybliżyć prawdopodobieństwo z gęstości zmiennej losowej za pomocą histogramu o bardzo małych szerokościach przedziałów. Użyjemy go do obliczenia prawdopodobieństwa, że ​​$P(X > 180cm)$.

```python
import numpy as np
import plotly.graph_objects as go
import utils

def student_normal(x):
    mean, std = 175, 4
    exponent = ((x-mean)/std)**2
    exponent *= -0.5
    output = np.exp(exponent)/(std*np.sqrt(2*np.pi))
    return output
```

```python
print(utils.prob_estimate(x_lower=180, x_upper=196, pdf_func=student_normal, bin_width=0.001))
utils.visualize_normal_interval(interval_floor=180, interval_ceil=196, normal_func=student_normal, mean=175, std=4)
```

Jak możemy użyć standardowego rozkładu normalnego, aby uzyskać tę samą odpowiedź?

Możemy ujednolicić nasz przedział w taki sposób, że:

$$P(X>x) \implies P(Z>\frac{x-\mu}{\sigma})$$
$$P(X>180) \implies P(Z > \frac{180-175}{4}) = P(Z > 1,25)$$

Gdzie 1,25 to odpowiedni _z-score_ lub _standard score_ surowego wyniku 180 cm. Możemy teraz zastosować tę samą metodę, jak to zrobiliśmy powyżej.

```python
def standard_normal(x):
    mean, std = 0, 1
    exponent = ((x-mean)/std)**2
    exponent *= -0.5
    output = np.exp(exponent)/(std*np.sqrt(2*np.pi))
    return output
```

```python
print(utils.prob_estimate(x_lower=1.25, x_upper=4, pdf_func=standard_normal, bin_width=0.001))
utils.visualize_normal_interval(interval_floor=1.25, interval_ceil=4, normal_func=standard_normal, mean=0, std=1, bw=0.01)
```

Zatem w obu przypadkach otrzymujemy takie same prawdopodobieństwa! Ale jak to działa? Cóż, ponieważ czynniki, które wpływają na prawdopodobieństwo to przesunięcie od średniej proporcjonalnie do wielkości odchylenia standardowego, __standardowy rozkład normalny zawiera wszystkie niezbędne informacje, biorąc pod uwagę, że mamy odpowiednie z-score.__ Oznacza to, że dla dowolnego przedziału dowolnego rozkładu normalnego, zawsze możemy użyć standaryzacji, aby spojrzeć na różne rozkłady normalne w tej samej strukturze. Innym powodem, dla którego warto mieć jedną holistyczną dystrybucję, jest to, że możemy sprawdzić __z-tabele.__

Te tabele zawierają __skumulowane prawdopodobieństwo__ standardowego rozkładu normalnego dla dowolnego podanego _z-score_ ($\Phi(z)$). Prawdopodobieństwo skumulowane mówi nam o prawdopodobieństwie $P(-\infty < Z < z)$ = $P(Z < z)$, a używając jednego lub wielu prawdopodobieństw skumulowanych z naszej tabeli z, możemy obliczyć prawdopodobieństwo dowolnego przedziału prawdopodobieństwa dowolnego rozkładu normalnego! Poniżej pokazuję, co oznaczają skumulowane prawdopodobieństwa z wizualizacją dla różnych wyników z.

```python
z_list = [-3, 0, 1.25, 3, 4]

for z in z_list:
    prob = utils.prob_estimate(x_lower=-4, x_upper=z, pdf_func=standard_normal, bin_width=0.001)
    print("cumulative_probability(z={z_val}) = {probability}".format(z_val=z, probability=prob))
    utils.visualize_normal_interval(interval_floor=-4, interval_ceil=z, normal_func=standard_normal, mean=0, std=1, bw=0.01)
    
```

Możemy potwierdzić te wyniki za pomocą konwencjonalnych tabel z dostępnych online, takich jak [tutaj](http://www.z-table.com/). Teraz możemy zobaczyć trzecią metodę obliczania prawdopodobieństwa $P(X>180)$. Metoda jest następująca:

$$P(X>180) = P(Z>1.25) $$
$$ = 1 - P(Z<1.25) $$
$$ = 1 - \Phi(1.25) $$
$$ = 1 - 0.8944 = 0.1056$$

Co daje nam ponownie to samo prawdopodobieństwo. Korzystając z takiego rozumienia standardowego rozkładu normalnego, możemy przeprowadzić tak zwane __z-testy__. Ten test wykorzystuje założenie normalności danych bazowych w celu określenia, czy wybrana hipoteza ma prawdopodobieństwo (wartość p) większe niż wybrany przez nas poziom istotności ($\alpha$). Możemy przybliżyć rozkład różnych ciągłych zmiennych losowych do rozkładów normalnych (pod warunkiem, że spełniają odpowiednie kryteria) lub zastosować _centralne twierdzenie graniczne (CLT)_ do porównania średnich różnych zestawów danych.

Rozważmy ten przykład: Netflix chce oficjalnie wprowadzić nową funkcję i chce sprawdzić, czy poprawi ona wrażenia użytkownika. Ich metodą ilościowego określania jakości doświadczenia użytkownika jest średni dzienny czas przesyłania strumieniowego dla ich bazy użytkowników. W związku z tym, patrząc na średni dzienny czas przesyłania strumieniowego danego użytkownika, Netflix może utworzyć __grupę kontrolną__ z użytkownikami niekorzystającymi z nowej funkcji oraz __grupę eksperymentalną__ z użytkownikami korzystającymi z nowej funkcji.

Ponieważ Netflix ma dostęp do średniego dziennego czasu transmisji każdego użytkownika, określili średnią populacji grupy kontrolnej na $\mu_{0} = 1,55 $ godzin. Z drugiej strony, aby określić skuteczność nowej funkcji, Netflix przeprowadza eksperyment na 100 użytkownikach, którzy są wprowadzani do nowej funkcji i zbierają średni dzienny czas transmisji. Jeśli możemy wykazać, że średnia populacji z nową funkcją jest __znacznie__ większa niż średnia populacji grupy kontrolnej, wykazaliśmy, że jest to __znacząca__ poprawa komfortu użytkowania. Przeprowadźmy teraz test hipotezy statystycznej, aby podjąć decyzję na podstawie zebranych danych, które importujemy jako ramkę danych poniżej.

```python
import pandas as pd
path = ''
stream_data = pd.read_csv(f"{path}/stream_data.csv")
```

Wykonując następujące kroki, możemy rozwiązać ten problem.

### 1. Sformułuj hipotezę
Biorąc pod uwagę naszą mentalność „niewinny o ile nie udowodnimy winy”, możemy zdefiniować nasze hipotezy zerowe i alternatywne w następujący sposób:

$\mathbf{H_{0}}$: Nowa funkcja nie ma wpływu na średni dzienny czas transmisji

$\mathbf{H_{a}}$: Nowa funkcja wydłuża średni dzienny czas transmisji

### 2. Znajdź odpowiedni test statystyczny
Ponieważ średnia próbki naszej grupy eksperymentalnej ma w przybliżeniu rozkład normalny (zakładając, że $n=100$), a wariancję populacji można aproksymować przez wariancję próbki, ponieważ $n$ jest duże (prawo dużych liczb), możemy przeprowadzić _z-test_.

### 3. Wybierz poziom istotności
Biorąc pod uwagę, że funkcja nie jest trudna do wdrożenia i może znacząco zwiększyć kapitał, nie musimy mieć ścisłego poziomu istotności. Dlatego użyjemy poziomu istotności $\alpha=0.05$.

### 4. Zbieraj dane i obliczaj statystyki testów
Ponieważ zebraliśmy już nasze dane, musimy obliczyć naszą statystykę testową, która w przypadku testu z jest znana jako __z-statystyka__. Zakładając, że nasza hipoteza zerowa jest prawdziwa, traktujemy średnią próbki jako próbkę z rozkładu normalnego podanego przez CLT (Centralne twierdzenie graniczne) z $\mu_{0}$ jako średnią populacji. Jeśli istnieje bardzo małe prawdopodobieństwo, że tak się stanie, możemy odrzucić naszą hipotezę zerową. Aby to obliczyć, oszacujemy również odchylenie standardowe próbki. Użyjemy Pythona do przeprowadzenia następującego procesu:

$$z = \frac{\bar{x} - \mu_{0}}{\frac{\sigma}{\sqrt{n}}} \approx \frac{\bar{x} - \mu_{0} }{\frac{s}{\sqrt{n}}} $$

__Uwaga__: Biorąc pod uwagę CLT, zakładamy, że średnia próbki naszych danych jest zgodna z rozkładem normalnym ze średnią populacji $\mu_{0}$ i wariancją równą stosunkowi wariancji naszej populacji do wielkości próby $\frac {\sigma^{2}}{n}$. Biorąc pod uwagę wystarczająco dużą wielkość próby, możemy bezpiecznie przybliżyć to odchylenie standardowe populacji, biorąc pod uwagę odchylenie standardowe próbki.

```python
import numpy as np

# Computing test statistic
mu_0 = 1.55
stream_time = stream_data["Average Stream Time (h)"]

sample_std = np.std(stream_time)
sample_mean = np.mean(stream_time)
n = stream_time.shape[0]

z_statistic = (sample_mean - mu_0)/(sample_std/(np.sqrt(n)))
print("z-statistic:", z_statistic)
```

### 5. Określ wartość p (prawdopodobieństwo)/wartość krytyczną
Możemy użyć naszej funkcji _prob\_estimate()_, internetowej tabeli z lub po prostu użyć biblioteki Pythona zawierającej tabelę z. Tabela z daje nam $p = 1 - 0,98 = 0,02 $. Poniżej przedstawiamy dwie pozostałe metody:

```python
import scipy.stats as stats

# Using our own prob_estimate function that uses histogram approximations
p1 = utils.prob_estimate(x_lower=z_statistic, x_upper=4, pdf_func=standard_normal, bin_width=0.001)
print("p1:", p1)

# Using scipy.stats
p2 = stats.norm.sf(z_statistic)
print("p2:", p2)

# Plotting the interval
utils.visualize_normal_interval(interval_floor=z_statistic, interval_ceil=4, normal_func=standard_normal, mean=0, std=1, bw=0.01)
    
```

### 6. Odrzuć/zaakceptuj hipotezę zerową
Biorąc pod uwagę, że nasza wartość p (0,02), która jest prawdopodobieństwem zaobserwowania średniej z naszej próbki lub czegokolwiek rzadszego przy normalnym rozkładzie wyśrodkowanym na $\mu_{0}$, jest znacznie mniejsza niż nasz poziom istotności, $\alpha$ ( 0,05), pozwala nam odrzucić hipotezę zerową, co oznacza, że ​​średnia próbki średniego dziennego czasu transmisji dla użytkowników nowej funkcji jest __znacznie__ większa niż średnia populacji użytkowników bez nowej funkcji.

Odstęp z pomarańczowym histogramem jest znany jako __obszar odrzucenia__ tego testu hipotezy i ma prawdopodobieństwo równe poziomowi istotności. Jeśli nasza statystyka testowa mieści się w tym obszarze, możemy odrzucić hipotezę zerową. Statystyka testu najbardziej zbliżona do średniej w obszarze odrzucenia jest znana jako __wartość krytyczna__. W tym przykładzie przeprowadzamy __test jednostronny__, ponieważ uwzględniamy efekt tylko w jednym kierunku. Wartość krytyczną dla testu jednostronnego z $\alpha=0.05$ można znaleźć w tabeli z, która wynosi $z_{0.05} = 1.65$. Ponieważ nasza statystyka z jest znacznie większa niż ta, znajduje się ona w obszarze odrzucenia i możemy odrzucić naszą hipotezę zerową. Jest to alternatywne podejście do odrzucania/nieodrzucania hipotezy zerowej.

### 7. Podejmij decyzję
Ponieważ funkcja ma statystycznie istotny wpływ na średnie użytkowanie produktu, wykorzystaliśmy nasze dane, aby obiektywnie zasugerować wdrożenie nowej funkcji!

To, co widzieliśmy, jest znane jako __test z jednej próby__, który określa, czy średnia próbki różni się statystycznie od znanej lub zakładanej średniej populacji. Możemy również użyć tej struktury w tak zwanych __dwu próbnych testach z__, aby określić, czy średnie różnice między dwiema różnymi populacjami są znacząco różne, czy też są wynikiem losowej szansy. Ze względu na CLT przydatne jest również użycie statystyki z dla __test z dwóch proporcji__.

# Testy T
Możemy teraz cofnąć się w czasie, aby dowiedzieć się o __teście t-studenta.__ W przeszłości testy Z były powszechne, ale były możliwe do wykonania tylko dla wystarczająco dużej próbki (zazwyczaj $n>30$). Jednak gdy wielkość próby była mała, wariancja próby nie była wystarczająco dobrym przybliżeniem wariancji populacji, i tu z pomocą przyszedł William Gosset. Był statystykem i piwowarem, który pracował dla Guinessa. Jego rolą było badanie i testowanie, które składniki zapewnią optymalny napar. Aby przezwyciężyć ograniczenia testu z dla średnich małych rozmiarów próbek, opracował __test t__. Aby Guiness zachował przewagę konkurencyjną, Gosset nie mógł otwarcie publikować swoich wyników. Dlatego swoje wyniki opublikował pod pseudonimem Student. To jest powód, dla którego nawet dzisiaj test ten nazywany jest testem t-Studenta!

Dlaczego więc potrzebujemy testu t i jak on działa? Jak wspomniano wcześniej, test z jest wiarygodny, o ile znana jest wariancja populacji (co jest wysoce nieprawdopodobne) lub jesteśmy w stanie oszacować ją na podstawie wariancji próbki, biorąc pod uwagę, że liczba obserwacji jest wystarczająco duża. Przyjrzyjmy się bliżej, jak to działa.

Wyobraź sobie, że pobieramy próbkę z ciągłej zmiennej losowej o rozkładzie wykładniczym ze średnią populacji $\mu=3$ i odchyleniem standardowym populacji $\sigma=3$. Obliczmy odchylenie standardowe próby dla różnych wielkości próby:

```python
n_vals = [2, 15, 50, 200] # different sample sizes to test

for n in n_vals:
    std_list = []
    for i in range(5):
        x = np.random.exponential(scale=3, size=n) # n observations
        std_list.append(np.std(x))
    print("n:", n, "sample standard deviations:", std_list)
```

Widzimy więc, że zgodnie z prawem wielkich liczb, nasze przybliżenie odchylenia standardowego populacji poprawia się wraz ze wzrostem liczby obserwacji. Jest to dodatkowo podkreślone przez fakt, że kiedy obliczamy statystykę z, odchylenie standardowe rozkładu jest podane przez wielkość $\frac{s}{\sqrt{n}}$. Jeśli poniżej znajdziemy tę ilość dla tych samych przypadków:

```python
for n in n_vals:
    std_list = []
    for i in range(5):
        x = np.random.exponential(scale=3, size=n) # n observations
        std_list.append(np.std(x)/np.sqrt(n))
    print("n:", n, "scaled stds:", std_list)
```

Oznacza to, że im większa liczba obserwacji, tym bardziej odpowiednie jest zastąpienie odchylenia standardowego populacji odchyleniem standardowym próby. A co, jeśli nie mamy dostępu do tylu obserwacji? Oznacza to, że odchylenie standardowe próbki nie może już przybliżać odchylenia standardowego populacji i samo jest teraz zmienną losową. Wynika to z faktu, że za każdym razem, gdy obliczyliśmy powyższe odchylenie standardowe, otrzymywaliśmy inną wartość. Dlatego __nie znamy wszystkich parametrów wymaganych do użycia rozkładu normalnego do testów statystycznych.__

Dlatego dla małych rozmiarów próbek (zazwyczaj $n<30$), zamiast obliczać statystykę z, obliczamy __t-statystykę__, aby przeprowadzić __t-test__ dla niezależnych identycznie rozłożonych próbek. Generalnie zakładamy, że podstawowy rozkład jest normalny, aby ten test zadziałał, podczas gdy według samego Williama Gosseta „wydaje się prawdopodobne, że odchylenie od normalności musi być bardzo ekstremalne, aby prowadzić do poważnego błędu”. Ta statystyka jest podana przez następujące równanie.

$$t_{n-1} = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$$

Który następuje po __rozkładzie t__ z tymi samymi parametrami, co odpowiadający mu standardowy rozkład normalny, ale z dodatkowym parametrem __stopnie swobody (df)__, gdzie $df = n-1$. Stopnie swobody w statystyce odnoszą się do liczby wartości w ostatecznych obliczeniach statystyki, które mogą się dowolnie zmieniać. Zobaczmy, jak wyglądają rozkłady t w odniesieniu do odpowiadającego im rozkładu normalnego.

```python
import plotly.graph_objects as go
import numpy as np
import scipy.stats as stats

dfs = [3, 10, 50]
x = np.linspace(-5, 5, 1000)

# Plotting different t-distributions
fig = go.Figure()
for df in dfs:
    rv = stats.t(df)
    fig.add_trace(go.Scatter(x=x, y=rv.pdf(x), name="t, df = {df}".format(df=df)))
fig.add_trace(go.Scatter(x=x, y=standard_normal(x), name="standard normal", line=dict(dash="dash")))
fig.update_layout(
    title="T-distributions",
    xaxis_title="x",
    yaxis_title="f(x)"
    )
```

Z wykresu widać, że rozkład t jest szerszą krzywą w kształcie dzwonu niż standardowy rozkład normalny, w którym parametr „stopni swobody” określa rozprzestrzenianie się gęstości zmiennej losowej. Ma to sens, ponieważ niższe $df$ implikuje większą niepewność, rozkład staje się szerszy. W związku z tym, gdy mamy coraz więcej obserwacji, rozkład t staje się węższy, zbliżając się do standardowego rozkładu normalnego. Dlatego nawet gdybyśmy użyli statystyki t i testu t, gdy mamy dużą liczbę próbek, otrzymalibyśmy takie same wartości p, jak w teście z!

Na przykład, gdybyśmy użyli testu t dla scenariusza Netflix, otrzymalibyśmy statystykę t $t=2.05$ i mielibyśmy $df=100-1 = 99$. Nie będziemy omawiać gęstości zmiennej losowej z rozkładami t w ramach tego wykładu ani przedmiotu, ale aby obliczyć prawidłową wartość p, możemy również użyć tabeli t dostępnej [online](http://www.ttable.org/ ) (który działa tak samo jak tabela z) lub wbudowany program w Pythonie. Z tabeli t otrzymujemy wartość p $p=0,02 $, tak jak widzieliśmy wcześniej!

```python
# p-value from t-distribution
t_statistic = 2.05
df = 100-1
p = stats.t.sf(abs(t_statistic), df=df)
print("p:", p)
```

### 3. Wybierz poziom istotności

### 4. Zbieraj dane i obliczaj statystyki testów
Ponieważ zebraliśmy już nasze dane, musimy obliczyć naszą statystykę testu, która w przypadku testu t jest znana jako __t-statystyka__. Użyjemy Pythona do przeprowadzenia następującego procesu:

$$t = \frac{\bar{x} - \mu_{0}}{\frac{s}{\sqrt{n}}} $$

# Testy chi-kwadrat

__Test chi-kwadrat__ jest testem statystycznym, który wykorzystuje rozkład __chi-kwadrat ($\mathbf{\chi^{2}}$)__ i zamiast tego obliczamy __statystykę chi-kwadrat.__ jest używany w innym kontekście niż testy z i t-testy. Jest niesamowicie potężny, ponieważ pozwala nam analizować relacje między dwiema kategorycznymi wielkościami.

Z definicji służy do określenia, czy istnieje statystycznie istotna różnica między licznościami oczekiwanymi a obserwowanymi w jednej lub kilku __wzajemnie wykluczających się__ kategoriach __tabeli kontyngencji.__ Tabela kontyngencji to macierz przechowująca rozkład liczności zmiennych. Jak działa test chi-kwadrat? Najpierw spójrzmy na przykład zabawkowego uczenia maszynowego z cechą wejściową „płeć” i wyjściową „ręczność”, odnosząc się odpowiednio do płci biologicznej osoby i dominującej ręki. Chcemy ustalić, czy wynik modelu uczenia maszynowego „ręczność” jest zależny od kategorii „płeć”. Jeśli jest niezależny, może to być funkcja, którą możemy usunąć z naszego zbioru danych.

|        | Praworęczny | Leworęczny | Razem |
|--------|--------------|-------------|-------|
| Mężczyzna | 43           | 9           | 52    |
| Kobieta | 44           | 4           | 48    |
| Razem  | 87           | 13          | 100   |

Jak na podstawie tych danych możemy wywnioskować jakiekolwiek informacje dotyczące związku między „płcią” a „ręcznością”? Cóż, jeśli te zmienne nie mają istotnego związku, wówczas spodziewalibyśmy się takiego samego stosunku częstości występowania mężczyzn do kobiet w grupie praworęcznej i leworęcznej. Przy takim założeniu możemy obliczyć oczekiwaną częstotliwość dla każdej __komórki__ w naszej tabeli kontyngencji. Pojawia się kluczowa idea testu chi-kwadrat: __jeśli obserwowane wartości różnią się zbytnio od wartości oczekiwanych, jeśli nie ma związku, musi istnieć jakiś związek między tymi dwiema zmiennymi.__ Dlatego możemy obliczyć statystykę chi-kwadrat następująco:

$$\chi^{2} = \sum_{i=1}^{n}\frac{(O_{i} - E_{i})^{2}}{E_{i}} $$

Sumując kwadrat różnic i skalując przez rozmiar oczekiwanych wartości każdej komórki, otrzymujemy ogólne odchylenie od przypadku, którego oczekiwalibyśmy, gdyby nie było związku. Teraz pojawia się pytanie: jak możemy określić ilościowo tę różnicę w ramach, które już widzieliśmy za pomocą testów hipotez? Z jakim rozkładem prawdopodobieństwa jest związana ta statystyka? Tutaj dzieje się magia statystyki. Jeśli możemy założyć, że same częstości mają __rozkład normalny__, to kwadrat chi jest sumą kwadratu $n$ zmiennych losowych o rozkładzie normalnym. W tym przypadku statystyka będzie zgodna z rozkładem chi-kwadrat zależnym od liczby komórek lub liczby wierszy i kolumn w naszej tabeli kontyngencji.

Obliczmy dla tego przykładu statystykę $\mathbf{\chi^{2}}$. Zaczniemy od określenia oczekiwanej wartości każdej komórki, jeśli te dwie zmienne są niezależne. Jeśli $\frac{52}{100} = 0,52 $ próby to mężczyźni, spodziewamy się, że 52% praworęcznych osób to mężczyźni. Dlatego możemy obliczyć oczekiwaną wartość w następujący sposób:

$$E_{RH-M} = 87 \times \frac{52}{100} = 45,24$$

Ta sama logika dotyczy oczekiwanej wartości innych komórek, jak pokazano poniżej:
$$E_{RH-F} = 87 \times \frac{48}{100} = 41,76$$
$$E_{LH-M} = 13 \times \frac{52}{100} = 6,76$$
$$E_{LH-F} = 13 \times \frac{48}{100} = 6,24$$

A teraz, gdy mamy oczekiwane wartości każdej komórki, możemy obliczyć naszą statystykę:

$$\chi^{2} = \sum_{i=1}^{n}\frac{(O_{i} - E_{i})^{2}}{E_{i}} = \frac{( 43 - 45,24)^2}{45,24} + \frac{(44 - 41,76)^2}{41,76} + \frac{(9-6,76)^{2}}{6,76} + \frac{(4-6,24) )^{2}}{6.24} = 1.777$$

Ta statystyka ma teraz rozkład chi-kwadrat! Aby określić dokładny rozkład, musimy znaleźć końcowy parametr dodatkowy, jakim są stopnie swobody. Powszechnym błędem jest założenie, że liczba stopni swobody jest równa liczbie komórek, ale tak nie jest. Ponieważ obliczenia wykorzystują sumy wierszy/kolumn, ostatnia komórka każdego wiersza/kolumny NIE może się zmieniać. Dlatego stopnie swobody można obliczyć w następujący sposób:

$$df = (\text{liczba wierszy} - 1)(\text{liczba kolumn} - 1)$$

Co w naszym przykładzie to po prostu $df=1$. Korzystając z naszych statystyk i $df$, możemy użyć __tabeli chi-square__, dostępnej [tutaj](https://www.medcalc.org/manual/chi-square-table.php) lub biblioteki Pythona zawierającej chi - dystrybucja kwadratowa. Poniżej przedstawiamy wykres zmian rozkładu chi-kwadrat w zależności od stopni swobody.

```python
# Plotting chi-square distributions
dfs = [1, 2, 5, 10]
x_vals = np.linspace(0.5, 8, 1000)

fig = go.Figure()
for df in dfs:
    fig.add_trace(go.Scatter(x=x_vals, y=stats.chi2.pdf(x_vals, df), name="df = " + str(df)))

fig.update_layout(
    title="Chi-square distributions",
    xaxis_title="x",
    yaxis_title="f(x)"
    )
fig.show()
```

Z powyższych rozkładów widać, że wraz ze zmianą liczby $df$ rozkład odpowiednio się zmienia, ponieważ oznacza to, że dodajemy do siebie coraz więcej kwadratów zmiennych losowych o rozkładzie normalnym. Rozważymy teraz ten przykład w ramach testowania hipotez, z którymi mieliśmy do czynienia wcześniej.

### 1. Sformułuj hipotezę
Jak wspomniano wcześniej, zakładamy najpierw, że dwie zmienne kategoryczne są niezależne:

$\mathbf{H_{0}}$: Te dwie funkcje są niezależne

$\mathbf{H_{a}}$: Te dwie funkcje nie są niezależne

### 2. Znajdź odpowiedni test statystyczny
Ponieważ mamy do czynienia ze zmiennymi kategorialnymi, użyjemy testu chi-kwadrat, aby określić, czy istnieje związek między tymi dwiema zmiennymi, czy też różnice w stosunku do oczekiwanych są wynikiem przypadku.

### 3. Wybierz poziom istotności
W tym zamierzonym celu będziemy kontynuować z poziomem istotności $\alpha=0.05$.

### 4. Zbieraj dane i obliczaj statystyki testów
Obliczyliśmy już statystykę chi-kwadrat, ale poniżej pokazujemy również, jak można ją obliczyć za pomocą bibliotek Pythona.

```python
# Przykład w Pythonie
obs = np.array([[43, 9], [44, 4]])
chi2, p_val,_,_ = stats.chi2_contingency(obs, correction=False)
print("chi-square:", chi2)
```

### 5. Określ wartość p (prawdopodobieństwo)/wartość krytyczną
Mając obliczoną statystykę i znane $df$, możemy użyć tabeli chi-kwadrat do określenia wartości krytycznej, która dla naszego poziomu istotności $\alpha=0.05$ wynosi $\chi^{2}_{C}= 3,84$. Możemy również po prostu użyć biblioteki Pythona do obliczenia odpowiedniej wartości p.

```python
# Wartość p wyliczona w komórce powyżej
print("p-value:", p_val)
```

### 6. Odrzuć/zaakceptuj hipotezę zerową
Biorąc pod uwagę, że nasza wartość p (0,182), która jest prawdopodobieństwem zaobserwowania naszej statystyki $\chi^{2}$, biorąc pod uwagę, że jest ona zgodna z rozkładem $\chi^{2}$, jest znacznie większa niż nasz poziom istotności, $\alpha=0.05$, nie możemy odrzucić hipotezy zerowej. Do tego samego wniosku można dojść, obserwując, że nasza obliczona statystyka jest mniejsza niż wartość krytyczna obserwowana w tabeli chi-kwadrat.

### 7. Podejmij decyzję
Nasze wyniki pokazują, że „ręczność” nie wykazuje znaczącej zależności od „płci”. W ten sposób możemy usunąć funkcję „płeć” z naszego modelu uczenia maszynowego bez znaczącego wpływu na jego dokładność.

To, co właśnie przeprowadziliśmy, jest znane jako __test niezależności chi-kwadrat__, który, jak sama nazwa wskazuje, służy do badania zależności jednej zmiennej kategorialnej od drugiej. Możemy również użyć __testów zgodności chi-kwadrat__, aby ocenić, na ile obserwowany rozkład naszych danych spełnia pewien oczekiwany rozkład dla wszystkich możliwych grup zmiennej kategorialnej.

# Testy F & ANOVA

Jak niektórzy mogli już się domyślić, biorąc pod uwagę kontekst tego tematu, __testy F__ są testami statystycznymi, które wykorzystują __statystyki F__, które są statystykami następującymi po __rozkładzie F__. Najczęściej testy F są przydatne podczas sprawdzania, czy średnia zbioru __populacji o rozkładzie normalnym__, z których wszystkie mają takie samo odchylenie standardowe, są równe. Odgrywa to istotną rolę w __analizie wariancji (ANOVA)__, która jest grupą modeli i procedur używanych do analizy różnicy między średnimi grup w próbie. Ta procedura może być postrzegana jako test statystyczny, który uogólnia test t poza dwie średnie. Przeanalizujemy tę zależność, przechodząc przez __jednoczynnikową ANOVA__.

Ta forma ANOVA jest używana, gdy masz kategoryczną zmienną niezależną (z co najmniej dwiema kategoriami) i zmienną zależną o rozkładzie normalnym (ciągłą) i chcesz __testować pod kątem różnic w średnich zmiennej zależnej w podziale na różne kategorie zmiennej niezależnej.__ W tym kontekście statystyka F jest obliczana w następujący sposób:

$$F = \frac{\text{wyjaśniona wariancja}}{\text{niewyjaśniona wariancja}} = \frac{\text{zmienność międzygrupowa}}{\text{zmienność wewnątrzgrupowa}}$$

W jaki sposób ta statystyka pomaga nam określić ilościowo różnice w średnich dla różnych grup? Rozważmy prosty zabawkowy przykład, czy rasa psa (kategoryczna zmienna niezależna) wpływa na wzrost psa (zmienna zależna od przedziału). Im __większa__ jest średnia różnica wzrostu (wariancja) w obrębie każdej rasy, tym __mniejsze prawdopodobieństwo__ występuje znacząca różnica w średniej populacji między różnymi rasami (grupami). Staje się to jeszcze bardziej zrozumiałe dzięki wykreśleniu różnych przypadków rozkładu wzrostu dla każdej rasy (zakładając, że zmienna ciągła ma rozkład normalny).

```python
# Analiza wizualizacji wariancji

from plotly.subplots import make_subplots

cases = ("High between, high within", "High between, low within", "Low between, high within", "Low between variance, low within")
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=cases)

mean_sets = [[40, 70, 100], [50, 60, 70]]
variances = [20, 5]
n_case = 0
x_vals = np.linspace(20, 120, 1000)
names = ["klee kai", "husky", "malamute"]
colors = ["orange", "black", "red"]

for idx1, mean_set in enumerate(mean_sets):
    for idx2, variance in enumerate(variances):
        for color, dog_name, mean in zip(colors, names, mean_set):
            current_pdf = stats.norm.pdf(x_vals, mean, variance**0.5)
            if idx1+idx2==0:
                fig.add_trace(go.Scatter(x=x_vals, y=current_pdf, marker_color=color, name=dog_name), row=idx1+1, col=idx2+1)
            else:
                fig.add_trace(go.Scatter(x=x_vals, y=current_pdf, marker_color=color, showlegend=False), row=idx1+1, col=idx2+1)
fig.show()            
```

Im bardziej rozkłady się pokrywają, tym trudniej jest odróżnić, czy średnie populacji każdej grupy są różne, jak pokazano powyżej. Jak więc obliczyć statystykę F?

Załóżmy, że nasza kategoryczna zmienna niezależna ma __I__ populacji (grup), a każda populacja ma __J__ obserwacji. Najpierw musimy obliczyć 3 wielkości pośrednie:

$$\text{Średnia z próbki populacji }i^{th}\text{: }\bar{X}_{i} = \frac{1}{J}\sum_{j=1}^{J }X_{ij} $$
$$\text{Przykładowa wariancja populacji }i^{th}\text{: }S_{i}^{2} = \frac{1}{J-1}\sum_{j=1}^{ J}(X_{ij} - \bar{X_{i}})^{2}$$
$$\text{Średnia z próby głównej: }\bar{X} = \frac{1}{I}\sum_{i=1}^{I}\bar{X_{i}} $$

Mając te zdefiniowane wielkości, możemy teraz obliczyć statystykę f. Ponieważ ANOVA jest powszechnym miejscem podczas analizy skuteczności leczenia medycznego na pacjentach, możemy obliczyć __średnia kwadratowa zabiegów (MSTr):__

$$MSTr = \frac{J}{I-1}\sum_{i=1}^{I}(\bar{X_{i}} - \bar{X})^{2}$$

Oprócz tego obliczamy błąd __średniokwadratowy__:

$$MSE = \frac{1}{I}\sum_{i=1}^{I}(S_{i}^{2})$$

W ostatnim kroku możemy obliczyć statystykę F, którą można wyrazić w dwóch postaciach:

$$F = \frac{MSTr}{MSE} = \frac{SS_{Leczenie}/(I-1)}{SS_{Błąd}/I(J-1)}$$

SS - suma kwadratów


Chociaż druga postać statystyki F niekoniecznie jest przydatna w obliczeniach, pokazuje ona, że ​​mianownik i licznik statystyki są podane przez sumę odpowiednich błędów kwadratowych, które są zgodne z rozkładem chi-kwadrat, podzieloną przez ich odpowiednie stopnie swobody. Innymi słowy, statystyka F jest zgodna z rozkładem określonym przez stosunek dwóch skalowanych zmiennych o rozkładzie chi-kwadrat! Dlatego ten rozkład F ma dwa parametry, którymi są __stopnie swobody licznika__ ($\mathbf{\nu_{1}}$) i __stopnie swobody mianownika__ ($\mathbf{\nu_{2}}$). Jak widzieliśmy wcześniej, stopnie swobody wariancji próbki z próbkami $N$ wynoszą $N-1$, ponieważ użycie średniej próby usuwa jeden stopień swobody ze statystyki. Z tą samą logiką widzimy, że MSTr musi mieć I-1 stopni swobody. Do obliczenia MSE mamy obserwacje I$\times$J i do obliczeń używamy średnich z próby grupy 'I'. Dlatego nasze stopnie swobody są podane w następujący sposób:

$$\nu_{1} = I-1$$
$$\nu_{2} = JI - I = I(J-1)$$

Zgodnie z naszym przykładem psa, jeśli mamy trzy grupy/populacje ($I=3$) i każda grupa ma 20 obserwacji ($J=20$), nasze stopnie swobody można obliczyć jako $\nu_{1 } = 3-1 = 2$ i $\nu_{2} = 20(3-1) = 40$. Poniżej przedstawiamy kilka przykładów rozkładów F dla różnych kombinacji stopni swobody.

```python
# f-distribution visualization

v1_list = [2, 2, 5, 5]
v2_list = [20, 40, 20, 40]
x_vals = np.linspace(0, 5, 1000)

fig = go.Figure()

for v1, v2 in zip(v1_list, v2_list):
    current_pdf = stats.f.pdf(x_vals, v1, v2)
    fig.add_trace(go.Scatter(x=x_vals, y=current_pdf, name="v1={v1}, v2={v2}".format(v1=v1, v2=v2)))

fig.update_layout(
    title="F-distributions",
    xaxis_title="x",
    yaxis_title="f(x)"
    )
fig.show()            
```

Przeprowadźmy teraz jednokierunkowy test ANOVA F, aby określić, czy średni wzrost populacji trzech różnych ras psów wykazuje znaczące różnice w naszej próbce.

### 1. Sformułuj hipotezę

$\mathbf{H_{0}}$: Średnia wysokość populacji jest równa dla każdej rasy psa

$\mathbf{H_{a}}$: Średnia wysokość populacji jest znacząco różna dla jednej lub więcej ras psów

### 2. Znajdź odpowiedni test statystyczny
Jak wspomniano, mając do czynienia z pojedynczą kategoryczną zmienną niezależną (rasa psa) i zmienną zależną od współczynnika, możemy użyć jednoczynnikowej analizy ANOVA.

### 3. Wybierz poziom istotności
Biorąc pod uwagę, że robimy to z ciekawości, nie będziemy tak skrupulatni jak wcześniej i wybierzemy poziom istotności $\alpha=0.10$.

### 4. Zbieraj dane i obliczaj statystyki testów
Poniżej wczytujemy dane dotyczące wzrostu psa i obliczamy statystykę f, stosując powyższe równania w Pythonie oraz korzystając z biblioteki _scipy.stats_.

```python
# Wczytanie danych
dog_data = pd.read_csv("../DATA/dog_data.csv")
dog_data = np.array([dog_data['kleekai'], dog_data['husky'], dog_data['malamute']]).T

# Zastosowanie równań
def MSTr(data):
    J, I = data.shape
    grand_mean = np.mean(data)
    group_mean_list = np.mean(data, axis=0)
    diffs = np.array([group_mean - grand_mean for group_mean in group_mean_list])
    MSTr = (J/(I-1))*np.sum(np.square(diffs))
    return MSTr

def MSE(data):
    J, I = data.shape
    group_mean_list = np.mean(data, axis=0)
    group_diffs = (1/(J-1))*np.sum(np.square(np.subtract(data, group_mean_list)), axis=0)
    MSE = (1/I)*np.sum(group_diffs)
    return MSE

def f_statistic(data):
    return MSTr(data)/MSE(data)

print("f:", f_statistic(dog_data))

# scipy.stats
print("f:", stats.f_oneway(dog_data[:, 0], dog_data[:, 1], dog_data[:, 2]))
```

### 5. Określ wartość p (prawdopodobieństwo)/wartość krytyczną
Mając obliczoną statystykę i znane parametry ($\nu_{1}=2$, $\nu_{2}=40$), możemy użyć [tabeli F](https://www.stat.purdue.edu/~jtroisi/STAT350Spring2015/tables/FTable.pdf) , aby określić wartość krytyczną, która dla naszego poziomu istotności $\alpha=0.1$ wynosi $F_{C}=2.84$. Obliczyliśmy również odpowiednią wartość p na podstawie zebranych danych, która wynosi $p=3.06\times10^{-20}$.

### 6. Odrzuć/zaakceptuj hipotezę zerową
Biorąc pod uwagę, że nasza wartość p , która jest prawdopodobieństwem zaobserwowania naszej statystyki F, biorąc pod uwagę rozkład f, jest znacznie niższa niż nasz poziom istotności $\alpha=0.1$, możemy odrzucić hipotezę zerową. Do tego samego wniosku można dojść, obserwując, że nasza obliczona statystyka ($F=109,39$) jest znacznie większa niż wartość krytyczna obserwowana w tabeli F.
