1. Korzystając z poniższego kodu oraz pliku `Zadanie_1.csv` utwórz tablicę zawierającą wzrost prezydentów USA.

````python
import numpy as np
import pandas as pd

tabela = pd.read_csv("Zadanie_1.csv")
tabela.head()
````

Korzystając z pakietu NumPy podaj:
- średni wzrost
- odchylenie standardowe
- najwyższy oraz najniższy wzrost:


````python
array = np.array(tabela["height(cm)"])
display(array)

mean_height = np.mean(array)
standard_deviation_height = np.std(array) 
max_height = np.max(array)
min_height = np.min(array)

print(f"Średni wzorst prezydentów to {mean_height}, odchylenie standardowe to {standard_deviation_height}, natomiast maksymalny wzrost to {max_height}, a minimalny {min_height}")
````

Podaj również 25 i 75 kwantyl oraz medianę:

````python
quantile25 = np.percentile(array, 25)
quantile75 = np.percentile(array, 75)
mediana = np.median(array)

print(f"kwantyl rzędu 25 to {quantile25}, rzędu 75 {quantile75}, natomiast mediana {mediana}")
````
